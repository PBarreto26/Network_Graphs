# Importando as bibliotecas

```python
import pandas as pd
import warnings
from pyvis.network import Network
warnings.filterwarnings('ignore')
```

# Ajustando o Índice

```python
df = pd.read_csv("main_names.csv", sep = ';')
```

```python
df['Indice'] = df['Unnamed: 0']
df = df.drop(columns='Unnamed: 0')
df
```

# Puxando uma base externa para fazer as relações de ganhadores do Oscar

```python
oscar = pd.read_csv('oscar.csv')
```

```python
nomes = []
indices = []
for i in df['Name']:
    p = df[df['Name'] == i]['Indice'].values
    if i in oscar[oscar['winner'] == True]['name'].unique():
        nomes.append(i)
        indices.append(p[0])
```

```python
for i in indices:
    nao_oscar.remove(i)
```

# Usando a base relações como aresta

```python
relations = pd.read_csv("relations.csv", sep = ';')
relations
```

```python
for i in nao_oscar:
    relations[0] = relations[relations[0] != i]
    relations[1] = relations[relations[1] != i]
```

```python
records = relations.to_records(index=False)
arestas = list(records)
```

```python
net = Network()
net.add_node(indices, label=nomes)
net.add_edges(arestas)
net.show('nodes.html')
```
